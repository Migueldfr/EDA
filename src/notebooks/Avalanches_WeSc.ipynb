```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = "https://www.avalanche.ca/incidents"
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
soup
```

```python
avalanches = soup.find_all('td')
avalanches
```

```python
date = soup.find('td').find('span').get_text(strip=True)
date
```

```python
dataset = [x.get_text(strip=True)for x in avalanches]
date = dataset[::8]
location = dataset[1::8]
province = dataset[2::8]
activity = dataset[3::8]
```

```python
involvement = [int(x.get_text(strip=True)) for x in avalanches[4::8]]
involvement
```

```python
injury = [x.get_text(strip=True) for x in avalanches][5::8]
injury
```

```python
fatal = [int(x.get_text(strip=True)) for x in avalanches[6::8]]
```

```python
avalanchesdata = {
    "Date": date,
    "Location": location,
    "Province": province,
    "Activity": activity,
    "Involvement": involvement,
    "Injury": injury,
    'Fatal' : fatal
}

results = pd.DataFrame(avalanchesdata)
```

```python
results['Injury'] = [x.replace('-','0') for x in results['Injury']]
```

```python
results['Injury']= pd.to_numeric(results['Injury'])
```

```python
#Estoy divideindo la columna fecha
date = results['Date'].str.split('-',expand=True)
date.columns = ['Year','Month','Day']
```

```python
#Estoy juntando los dos dataframe para operar con ellos
avalanchesfatalities = pd.concat([results,date],axis=1)
avalanchesfatalities.drop(['Day','Date'], axis=1, inplace=True)
```

```python
def mes(row):
    '''
    Esta funcion es para determinar el numero de mes que es y cambiarlo por el nombre de mes para poder hacerlo
    mas visual y facil de leer.
    '''
    if row['Month'] == '01':
        row['Month'] = 'January'
    elif row['Month'] == '02':
        row['Month'] = 'February'
    elif row['Month'] == '03':
        row['Month'] = 'March'
    elif row['Month'] == '04':
        row['Month'] = 'April'
    elif row['Month'] == '05':
        row['Month'] = 'May'
    elif row['Month'] == '06':
        row['Month'] = 'June'
    elif row['Month'] == '07':
        row['Month'] = 'July'
    elif row['Month'] == '08':
        row['Month'] = 'August'
    elif row['Month'] == '09':
        row['Month'] = 'September'
    elif row['Month'] == '10':
        row['Month'] = 'October'
    elif row['Month'] == '11':
        row['Month'] = 'November'
    elif row['Month'] == '12':
        row['Month'] ='December'
        
    return row['Month']

avalanchesfatalities['Month'] = avalanchesfatalities.apply(mes, axis=1)
```
