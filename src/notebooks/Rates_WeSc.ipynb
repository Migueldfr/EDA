```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option("max_rows", 1000)
```

```python
url = 'https://www.skiresort.info/best-ski-resorts/europe/page/4/'
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
print(soup)
```

```python
rate = soup.find('div', id='resortList').find_all('div', class_='rating-list js-star-ranking stars-middle')
rate
```

```python
rates = [float(x.get('data-rank')) for x in rate]
print(rates)
```

```python
link1 = soup.find('div', id='resortList').find_all('div', class_='h3')
link1
```

```python
link2 = [x.find('a').get('href')[:-12] for x in link1]
link2
```

```python
links = []
for i in range(len(link2)):
    link= link2[i][:4]+link2[i][5:]
    links.append(link)
links
	
```

```python
len(links)
```

```python
print(len(rates))
print(len(links))
```

```python
x = {'URL':links,
        'Rate':rates}
```
